# Visualization: Example using pyvista

```python
import pyvista
usingJupyterNotebook = True
```

## Option 1: Direct plot

```python
sol = pyvista.read('tutorial-fem/x_solution.vtu')
sol.plot(notebook=usingJupyterNotebook, interactive=True)
```

## Option 2: Fine-grained configuration using Plotter

```python
plotter = pyvista.Plotter(2,2)
plotter.add_mesh(sol, scalars="u", clim=[0, 1], cmap="fire", lighting=True, stitle="Solution u")
plotter.show()
```
